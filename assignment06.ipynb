## import packages 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define variable & function

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b
```

```python
fp1 = np.polyfit(x, y1, 1)
f1 = np.poly1d(fp1)
```

## Plot the noisy data (x, y1)

```python
plt.figure(figsize=(6,4))
plt.plot(x, y1, label='noise', color='y')
```

## Plot the clean data (x, y2)

```python
plt.figure(figsize=(6,4))
plt.plot(x, y2, ls='dashed', lw=3, color='b', label='clean')
```

## Plot the line that fits the noisy data by the least square error

```python
plt.figure(figsize=(6,4))
plt.plot(x, f1(x), lw=2, color='r', label='polyfit')
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, y1, label='noise', color='y')
plt.plot(x, y2, ls='dashed', lw=3, color='b', label='clean')
plt.plot(x, f1(x), lw=2, color='r', label='polyfit')
plt.grid()
plt.legend()
plt.show()
```
